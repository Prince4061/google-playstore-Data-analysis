# **Project Summary -**

This project aimed to identify the key attributes of successful mobile apps on the Google Play Store by analyzing a dataset containing app information and user reviews. Through exploratory data analysis (EDA), we examined various aspects such as app ratings, the popularity of different app categories, the prevalence of free versus paid apps, the relationships between numerical variables like ratings, reviews, installs, and price, and the sentiment expressed in user reviews.

Key Findings

*   **High App Ratings:** Successful apps tend to have high ratings, primarily concentrated between 4.0 and 4.5.
*   **Competitive Categories:** Categories like Family and Games are highly competitive.
*   **Dominance of Free Apps:** The market is largely dominated by free apps, indicating effective alternative monetization strategies.
*   **Correlation Between Reviews and Installs:** There is a positive correlation between the number of reviews and app installs, highlighting the importance of user feedback.
*   **User Sentiment:** Analysis of user reviews revealed insights into overall sentiment distribution and how sentiment varies across different app categories and content ratings. Positive sentiment is prevalent, but negative and neutral reviews highlight areas for improvement.

Recommendations for App Developers

Based on these insights, here are some recommendations for developers to enhance their chances of success:

*   **Prioritize App Quality:** Focus on improving app quality and user experience to achieve and maintain high ratings.
*   **Strategic Category Navigation:** Be strategic about choosing popular and competitive categories, considering market saturation and identifying niches.
*   **Consider Freemium Models:** Explore freemium models to attract a larger user base, given the dominance of free apps.
*   **Encourage User Reviews and Address Sentiment:** Actively seek user reviews and, importantly, analyze and respond to feedback, particularly addressing negative sentiment to improve user satisfaction and boost credibility.
*   **Tailor Strategy Based on Sentiment by Category and Content Rating:** Utilize the insights on sentiment variation across categories and content ratings to tailor app development, marketing, and content strategies for better resonance with target audiences.

## 3. ***Data Wrangling***

````python
# Check for unique values in the Price column
df['Price'].unique()
````

````python
# Checking the row number where the Column Name is 'Everyone'
df[df['Price'] == 'Everyone']
````

````python
# Removing row 10472 because it contains a data entry error.
df.drop(10472,inplace=True)
````

````python
# Checking the unique values in the 'Price' column.
df['Price'].unique()
````

### Data Wrangling Code

````python
# Clean the 'Price' column: remove '$' and convert to float.
df['Price'] = df['Price'].str.replace('$','').astype(float)
````

````python
# checking unique values from the Price column
df['Price'].unique()
````

````python
# check unique values from the installs column
df['Installs'].unique()
````

````python
# Clean 'installs' column: remove '+' and ',' characters and convert to integer.
df['Installs'] = df['Installs'].str.replace('+','').str.replace(',','')
````

````python
# Checking unique values in 'installs' column after cleaning.
df['Installs'].unique()
````

````python
# Convert install column to a numeric value.
df['Installs'] = df['Installs'].astype(int)
````

````python
# check unique values in the 'rating' column
df['Rating'].unique()
````

````python
# Fill the null values in the rating column with the median.
df['Rating'] = df['Rating'].fillna(df['Rating'].median())
````

````python
# Converting the 'rating' column to a numeric type.
df['Rating'] = df['Rating'].astype(float)
````

````python
# changing Removing M , K in size column and converting it into MB
df['Size'] = df['Size'].str.replace('M','000')
df['Size'] = df['Size'].str.replace('k','')
df['Size'] = df['Size'].replace('Varies with device',np.nan)
df['Size'] = df['Size'].astype(float)
df['Size'] = df['Size'].apply(lambda x: x/1000 if x > 100 else x)
````

````python
# Final check for any remaining missing values in the DataFrame.
df.isnull().sum()
````

````python
# Dropping rows with any missing values.
df.dropna(inplace=True)
````

# Checking the missing values again.


````python
df.isnull().sum()
````

````python
# Merge the two dataframes on the 'App' column
merged_df = pd.merge(df, df_review, on='App', how='inner')

# Display the first few rows of the merged dataframe
display(merged_df.head())

# Display the shape of the merged dataframe
print("\nShape of the merged dataframe:", merged_df.shape)
````

### What all manipulations have you done and insights you found?

* Dropped the row where 'Price' was 'Everyone'.
* Removed the '$' symbol from the 'Price' column and converted it to a float data type.
* Filled the missing values in the 'Rating' column with the median rating.
* Converted the 'Installs' column to a numeric data type, coercing errors to NaN, and then dropped rows with NaN values in this column.
* Removed the remaining rows with missing values.
